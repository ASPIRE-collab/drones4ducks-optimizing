```python
import pandas as pd

path = "E:/imagefactors/data/crowdsourced_imagefactors.csv"
with open(path) as f:
  df = pd.read_csv(f)

file_list = df['basefile'].unique().tolist()
len(file_list)
```

```python
import os
import shutil

# Define the source directory containing images
source_directory = 'Y:\\GEMdata\\USFWS\\DronesForDucks\\Data\\DronesForDucksDatabase\\Data\\Imagery\\20220129_RGNC\\0_Originals'

# Define the destination directory to copy matching images
destination_directory = 'E:\\imagefactors\\data\\originals'

# Create the destination directory if it doesn't exist
if not os.path.exists(destination_directory):
    os.makedirs(destination_directory)

# Iterate over the list of filenames
for filename_without_extension in file_list:
    # Iterate over common image file extensions (you can add more if needed)
    for extension in ['.jpg', '.png', '.jpeg']:
        # Construct the full path to the source image
        source_image_path = os.path.join(source_directory, f'{filename_without_extension}{extension}')
        
        # Check if the source image exists
        if os.path.exists(source_image_path):
            # Construct the full path to the destination image
            destination_image_path = os.path.join(destination_directory, f'{filename_without_extension}{extension}')
            
            # Copy the source image to the destination directory
            shutil.copy(source_image_path, destination_image_path)
            print(f'Copied {filename_without_extension}{extension} to {destination_directory}')
            break  # Exit the loop after the first match is found
    
    else:
        print(f'{filename_without_extension} not found in {source_directory}')

print('Copy operation completed.')
```
