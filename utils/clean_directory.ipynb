```python
#Remove images with no corresponding label-- this deletes them, so make sure there is a backup!

import os

def remove_images_without_labels(image_dir, label_dir):
    for split in ['train', 'val', 'test']:
        image_split_dir = os.path.join(image_dir, split)
        label_split_dir = os.path.join(label_dir, split)
        
        image_files = os.listdir(image_split_dir)
        label_files = os.listdir(label_split_dir)
        
        for image_file in image_files:
            image_path = os.path.join(image_split_dir, image_file)
            label_file = image_file.replace('.png', '.txt')  # Assuming labels have the same name as images but with .txt extension
            label_path = os.path.join(label_split_dir, label_file)
            
            if not os.path.exists(label_path):
                os.remove(image_path)
                print(f"Removed image: {image_path}")
```

```python
image_dir = 'E:/datasets/zooniverse_01/images/'
label_dir = 'E:/datasets/zooniverse_01/labels/'

remove_images_without_labels(image_dir, label_dir)
```
